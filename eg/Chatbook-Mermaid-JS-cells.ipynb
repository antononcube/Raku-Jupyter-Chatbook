## Mermaid-JS specs

It is more correctly to say that we produces images from [Mermaid-JS](https://mermaid.js.org) specs via [mermaid-ink](https://mermaid.ink).

"Under the hood" the Raku package ["WWW::MermaidInk"](https://raku.land/zef:antononcube/WWW::MermaidInk) is utilized.

Doing it "manually":

```raku
use WWW::MermaidInk;
use Text::Plot;

my $spec = q:to/END/;
graph LR
    A --> B --> C
    C --> D --> B
END

mermaid-ink($spec, format => 'md-image') ==> from-base64
```

Rendering "directly" from specification:

```raku
%%mermaid
graph LR
    A --> B --> C
    C --> D --> B
```

Bigger spec of a sequence diagram:

```raku
%%mmd
sequenceDiagram
  participant User
  participant llmfunc as llm-function
  participant llmconf as llm-configuration
  participant LLMConf as LLM configuration
  participant LLMEval as LLM evaluator
  participant AnonFunc as Anonymous function
  User ->> llmfunc: prompt<br>conf spec
  llmfunc ->> llmconf: conf spec
  llmconf ->> LLMConf: conf spec
  LLMConf ->> LLMEval: wrap with
  LLMEval ->> llmfunc: evaluator object
  llmfunc ->> AnonFunc:  create with:<br>evaluator object<br>prompt
  AnonFunc ->> llmfunc: handle
  llmfunc ->> User: handle
```
